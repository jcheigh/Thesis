```python
n = 10

def decimal_to_binary(n):
    """
    returns binary string representing integer n (n >= 0)
    """
    if n == 0:
        return ''
    
    else:
        return decimal_to_binary(n // 2) + f'{n % 2}'

def decimal_to_baseB(n, b):
    """
    returns base B string representing integer n (n >= 0)
    """
    if n == 0:
        return ''
    
    else:
        return decimal_to_baseB(n // b, b) + f'{n % b}'

def binary_to_decimal(bin_str):
    if len(bin_str) == 1:
        return int(bin_str)
    
    return int(bin_str[-1]) + 2 * binary_to_decimal(bin_str[:-1])

def baseB_to_decimal(b_str, b):
    if len(b_str) == 1:
        return int(b_str)
    
    return int(b_str[-1]) + b * baseB_to_decimal(b_str[:-1], b)

def baseA_to_baseB(n, a, b):
    decimal = baseB_to_decimal(str(n), a)
    return decimal_to_baseB(int(decimal), b)

for n in range(1, 5000):
    for b in range(2, 10):
        assert n == baseB_to_decimal(decimal_to_baseB(n, b), b)
```

```python
b1 = 10; b2 = 4

print(f'b1: {decimal_to_binary(b1)}, b2: {decimal_to_binary(b2)}')

print(f'Bitwise And: {decimal_to_binary(b1 & b2)}')
print(f'Bitwise OR: {decimal_to_binary(b1 | b2)}')
print(f'Bitwise XOR: {decimal_to_binary(b1 ^ b2)}')
print(f'Bitwise Not: {bin(~b1)}')
```

```python
def bitwise_and(n1, n2):
    bin_n1 = decimal_to_binary(n1)
    bin_n2 = decimal_to_binary(n2)

    length1 = len(bin_n1)
    length2 = len(bin_n2)

    if length1 >= length2:
        result = '0' * (length1 - length2)
        bin_n1 = bin_n1[length1 - length2:]
    else:
        result = '0' * (length2 - length1)
        bin_n2 = bin_n2[length2 - length1:]

    for b1, b2 in zip(bin_n1, bin_n2):
        result += '1' if b1 == b2 == '1' else '0'
    
    return result

def bitwise_or(n1, n2):
    bin_n1 = decimal_to_binary(n1)
    bin_n2 = decimal_to_binary(n2)

    length1 = len(bin_n1)
    length2 = len(bin_n2)

    if length1 >= length2:
        result = '0' * (length1 - length2)
        bin_n1 = bin_n1[length1 - length2:]
    else:
        result = '0' * (length2 - length1)
        bin_n2 = bin_n2[length2 - length1:]

    for b1, b2 in zip(bin_n1, bin_n2):
        result += '1' if '1' in [b1, b2] else '0'
    
    return result

def bitwise_xor(n1, n2):
    bin_n1 = decimal_to_binary(n1)
    bin_n2 = decimal_to_binary(n2)

    length1 = len(bin_n1)
    length2 = len(bin_n2)

    if length1 >= length2:
        result = '0' * (length1 - length2)
        bin_n1 = bin_n1[length1 - length2:]
    else:
        result = '0' * (length2 - length1)
        bin_n2 = bin_n2[length2 - length1:]

    for b1, b2 in zip(bin_n1, bin_n2):
        result += '1' if (b1, b2) in [('1', '0'), ('0', '1')] else '0'
    
    return result
```
